# Predicting House Prices with XGBoost on California Dataset

This project focuses on predicting house prices using the California Housing Dataset available in sklearn.datasets. We apply the XGBoost Regressor, which is well-known for its high performance and efficiency, especially in tabular data. The objective is to train the model, evaluate its performance, and analyze prediction results using visualizations and metrics.

#Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

#Importing the California House Price Dataset

```python
house_price_dataset=sklearn.datasets.fetch_california_housing()
```

```python
house_price_dataset
```

```python
#loading the dataset into a pandas dataframe
house_price_dataframe=pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe
```

```python
#adding the (price) target column
house_price_dataframe['price']=house_price_dataset.target
```

```python
#checking the first five rows of the dataframe
house_price_dataframe.head()
```

```python
#checking the no. of rows and columns of the dataframe
house_price_dataframe.shape
```

```python
#checking the missing values
house_price_dataframe.isnull().sum()
```

```python
#statestical measures of the dataset
house_price_dataframe.describe()
```

#Understanding the correlation between the various featues of the dataset

1.Positive Correlation

2.Negative Correlation

```python
correlation=house_price_dataframe.corr()
correlation
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

#Splitting the Data and the Target

```python
X=house_price_dataframe.drop("price",axis=1)
Y=house_price_dataframe.price
```

```python
X,Y
```

#Splitting the Data into Train And Test Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X.shape,X_train.shape,X_test.shape
```

#Model Training

##XGBoost Regressor

```python
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

#Prediction on training data and Evaluation

```python
X_train_prediction=model.predict(X_train)
```

```python
#R squared error
score1=metrics.r2_score(X_train_prediction,Y_train)
#MAE
score2=metrics.mean_absolute_error(X_train_prediction,Y_train)
```

```python
print("R squared error: ",score1)
print("Mean Absolute Error: ",score2)
```

#Visualizing the Actual and Predicted Price

```python
plt.scatter(Y_train,X_train_prediction,alpha=0.05)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual VS Predicted Prices")
```

#Prediction on test data and Evalution

```python
X_test_prediction=model.predict(X_test)
```

```python
#R Squared Error
score1=metrics.r2_score(X_test_prediction,Y_test)
#MAE
score2=metrics.mean_absolute_error(X_test_prediction,Y_test)
```

```python
print("R squared error: ",score1)
print("Mean Absolute Error: ",score2)
```

#Visualizing the Actual and Predicted Prices

```python
plt.scatter(Y_test,X_test_prediction,alpha=0.05)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual VS Predicted Prices")
```

# Conclusion

This project demonstrates the effectiveness of XGBoost in predicting housing prices with strong results on both training and testing data. To improve the model, future work could involve hyperparameter tuning, feature engineering, and applying cross-validation for robustness.
